## List Slicing


```python
fruits = ["Apple", "Banana", "Cherry", "Orange", "Mango", "Pineapple"]
```


<div style="display: flex; gap: 20px; align-items: flex-start;">

<div>

| List Items        | Apple | Banana | Cherry | Orange | Mango | Pineapple |
|-------------------|-------|--------|--------|--------|-------|-----------|
| **Position**      | 1     | 2      | 3      | 4      | 5     | 6         |
| **Positive Index**| 0     | 1      | 2      | 3      | 4     | 5         |
| **Negative Index**| -6    | -5     | -4     | -3     | -2    | -1        |


</div>

</div>

#### Slicing direction
- list[start: stop: step]
- positive step → slicing direction: left to right → (start position < stop position)
- negative step → slicing direction: right to left → (start postion > stop position)

#### Default start, end, step values

- default step = 1

- positive step
  - default start = 0
  - default end = len(list)
- negative step
  - default start = -1
  - default end = - ( len(list) + 1 )

#### 📌 Most important slicing patterns
- list[start:end] → extracts elements starting at `start` up to just before `end`.

- list[start:] → extracts elements from `start` all the way to the last item.

- list[:end] → extracts elements from the first item up to just before `end`.

- list[::-1] → returns the list in reverse order.

````python
fruits = ["Apple", "Banana", "Cherry", "Orange", "Mango", "Pineapple"]
# position 2-5
print(fruits[1:4:])
print(fruits[-5:-2])
print(fruits[1:-2])
````

````python
fruits = ["Apple", "Banana", "Cherry", "Orange", "Mango", "Pineapple"]
# position 2-5
print(fruits[3:0:-1])
print(fruits[-3:-6:-1])
print(fruits[-3:0:-1])
````

````python
fruits = ["Apple", "Banana", "Cherry", "Orange", "Mango", "Pineapple"]
fruits[::-2]
````

````python
# Most important patterns
fruits = ["Apple", "Banana", "Cherry", "Orange", "Mango", "Pineapple"]
print(fruits[2:5])
print(fruits[3:])
print(fruits[:3])
print(fruits[::-1])
````

````python
# Think of positions, not the index
# start index = 2 (postive index) =  -4 (negative index) = 3rd (position) = "Cherry"
# end index = 5 (positive index) = -1 (negative index) = 6th/last (position) = "Pineapple"
# Notice that here step is positive. So, (start position e.g. 3rd) < (end position e.g. 6th)
# slicing direction is left to right


fruits = ["Apple", "Banana", "Cherry", "Orange", "Mango", "Pineapple"]
print(fruits[2:5:1])
print(fruits[-4:-1:1])
print(fruits[-4:5:1])
````

````python
# Think of positions, not the index
# start index = 4 (postive index) =  -2 (negative index) = 5th (position) = "Mango"
# end index = 1 (positive index) = -5 (negative index) = 2nd (position) = "Banana"
# Notice that here step is negative. So, (start position e.g. 5th) > (end position e.g. 2nd) and
# slicing direction is right to left

fruits = ["Apple", "Banana", "Cherry", "Orange", "Mango", "Pineapple"]
print(fruits[4:1:-1])
print(fruits[-2:-5:-1])
print(fruits[4:-5:-1])
````

## `Tuple` intro and creation

A `tuple` in Python is an `ordered`, `immutable` (unchangeable) collection of items. Think of it as a "Locked List."
Once you write data into it, no one can change it, delete it, or rearrange it.

🌐 Real-World Use Case: Database Records 📊

When you fetch a user’s birthdate or a transaction ID from a database, you store it in a tuple to ensure the code doesn't accidentally change a customer's ID during a calculation.

````python
t1 = (1, 2, 3)
t2 = ("Azmain", 1, True, 4.5, "a")
t3 = (1, 2, 3, ("a", "b"))
t4 = (1, 2, 3, ("a", "b"), [1,2,7])
````

## Mutability vs Immutability

- Mutability refers to the ability of an object to be changed or modified after creation.
- Mutable data types → can be changed after creation → `list`, `set`, `dict`
- Immutable data types → cannot be changed after creation → `tuple`, `int`, `float`, `str`, `bool`

````python
# list is mutable / modifiable after creation

lst = [1,2, "Hello", "World", 10]
print(lst)
lst[3] = "Bangladesh"     # ✔ Correct (List is mutable)
print(lst)
````

````python
# tuple is immutable / unmodifiable after creation

t = (1,2, "Hello", "World", 10)
print(t)
t[3] = "Bangladesh"    # ❌ Error (Tuple is immutable)
````

````python
t = [1,2, "Hello", "World", 10]
print(t)
````

## `Set` Intro and Creation
A `set` in Python is an `unordered` collection of `unique` elements


🌐 Real-World Use Case: Digital Marketing & NLP 🛒📈

If you have a list of 10,000 email subscribers and you buy another list, you don't want to send duplicate emails to the same person. Sets handle this instantly.


````python
s1 = {1, 3, 2}
s2 = {"apple", "banana", "mango"}
s3 = {"hello", 1, 3.14, 1}
s4 = {1, "hey", "b", "hey", 1, "hey"}

# notice the order and uniqueness of set elements after printing the sets
print(s1)
print(s2)
print(s3)
print(s4)
````

````python
s5 = set(range(1,20,2)) # set constructor
print(s5)
````

````python
# Note: Creating an Empty Set


s1 = {}
s2 = set()

print(type(s1))  # <class 'dict'>
print(type(s2))  # <class 'set'>

# `{ }` always creates an empty dictionary, not a set

# An empty set can only be created using the `set()` constructor
````

## Set Methods

####  `Set`: Adding and Removing elements

**Note:**

- `remove()` → Raises a KeyError if the element is not present in the set

- `discard()` → Does nothing if the element is not present (no error raised)

```python
s = {1, 2, 3}

s.remove(2)    # removes 2 from set
s.remove(4)    # ❌ KeyError
s.discard(4)   # ✅ No error, safe removal
```

#### Set Operations (using methods)
```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A.union(B))          # Union → all unique elements from both sets (common + uncommon)
print(A.intersection(B))   # Intersection → elements common to both sets
print(A.difference(B))     # Difference → elements present only in A
print(B.difference(A))     # Difference → elements present only in B
```
Output:
```shell
{1, 2, 3, 4, 5, 6, 7, 8}
{4, 5}
{1, 2, 3}
{8, 6, 7}
```

#### Set Operations (using operators)

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

print(A | B)   # Union → all unique elements from both sets (common + uncommon)
print(A & B)   # Intersection → elements common to both sets
print(A - B)   # Difference → elements present only in A
print(B - A)   # Difference → elements present only in B
```
Output:
```shell
{1, 2, 3, 4, 5, 6, 7, 8}
{4, 5}
{1, 2, 3}
{8, 6, 7}
```

````python
old_subscriber = {"rafin@gmail.com", "shafin@gmail.com", "sadia@gmail.com", "hasan@gmail.com"}
new_subscriber = {"sadia@gmail.com", "hasan@gmail.com", "sayem@gmail.com", "adnan@gmail.com"}

# Total unique subscribers
total_unique = old_subscriber.union(new_subscriber)
print("Total Unique Subscribers:", total_unique)

# Users who shifted to new channel (common subscribers)
shifted_users = old_subscriber.intersection(new_subscriber)
print("Users shifted to new channel:", shifted_users)

# Users who stayed only in old channel
only_old = old_subscriber.difference(new_subscriber)
print("Only in old channel:", only_old)

# Users who are completely new
only_new = new_subscriber.difference(old_subscriber)
print("Only in new channel:", only_new)
````

````python
## Set fast membership check

import timeit

# Create large data
n = 1_000_000
lst = list(range(n))
tpl = tuple(range(n))
st = set(range(n))

# Membership test for last element
list_time = timeit.timeit("999999 in lst", globals=globals(), number=1000)
tuple_time = timeit.timeit("999999 in tpl", globals=globals(), number=1000)
set_time = timeit.timeit("999999 in st", globals=globals(), number=1000)

print(f"List  : {list_time} ~ {list_time:.02f} seconds")
print(f"Tuple : {tuple_time} ~ {tuple_time:.02f} seconds")
print(f"Set   : {set_time} ~ {set_time*100000:.02f} microseconds ")
````

## `Dictionary` intro and creation

A dictionary in Python is a built-in, `mutable`, and `ordered` collection data type used to store data in `key-value` pairs.

🌐 Real-World Use Case: Search Engine 🔎

Think of a dictionary like a search engine:

- You type a keyword (key)

- The search engine returns relevant results (value)

````python
d1 = {"name": "Saihan", "age": 20, "work": "Ostad"}
print(d1)
````

````python
d2 = dict(name="Hasan", age=26, city="Dhaka")  # dictionary constructor function
print(d2)
````

````python
d3 = {"a": 10, "b": 20, "c": 30}
d4 = {"a": "abc", "b": "bcd", "c": "cdf"}
d5 = {"a": True, "b": 45, "c": [1,2,3], "d": (7,9,11), "e": {"x": 10, "y": 30}}
d6 = {1: "abc", 2:"bcd", 3:"cdf", 3.14:"pi"}
````

````python
d7 = {
    "name": "Hasib",
    "city": "Chittagong",
    "city": "Dhaka"
}
print(d7)
````

## `Dictionary` read/access values
- Unlike lists or tuples, dictionaries do not support indexing. You cannot access elements by position; you must use keys to retrieve values.

```python

print(data["name"])   # ✅ key-based access
print(data[0])        # ❌ KeyError (no index-based access)
```
