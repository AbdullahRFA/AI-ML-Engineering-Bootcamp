Question: Online Shopping Cart System (30 Marks)

An e-commerce website wants to develop a basic shopping cart system using Object-Oriented Programming (OOP) in Python.

Each shopping cart belongs to a customer and stores multiple item prices. The system should calculate totals, apply discounts, display cart details, and allow cart combinations.

# **System Requirements**
#### **Each shopping cart must store:**

- Customer name

- A list of item prices added to the cart

**The system should be able to:**

- Calculate total bill amount

- Apply discount rules

- Display cart details

- Combine two shopping carts

- Perform basic cart operations such as adding and removing items

# **Part A – Class Creation (7 Marks)**
Create a class named ShoppingCart with the following:

Attributes:

- customer_name → Name of the customer (string)

- items → List of item prices (list of numbers)

**Constructor:**

- Use init() to initialize all attributes correctly

- If no items are provided, initialize with an empty list

# **Part B – Core Methods Implementation (12 Marks)**
Implement the following methods inside the **ShoppingCart class:**

1. **add_item(price)**

- Adds a new item price to the cart

- Price must be a positive number

2. **remove_item(price)**

- Removes one occurrence of the given price from the cart

- If the item does not exist, display a suitable message

3. **calculate_total()**

- Uses a loop or list comprehension

- Returns the total price of all items in the cart

4. **apply_discount()**

**Discount rule:**

➤Total ≥ 3000 → apply 10% discount

➤Total < 3000 → no discount

- Returns the final payable amount

# **Part C – Display & Utility Methods (6 Marks)**
1. **display_cart()**

Prints the following in a well-formatted manner:

- Customer name

- Items in the cart

- Total amount

- Final amount after discount

- Number of items in the cart

2. **clear_cart()**

- Removes all items from the cart

- Displays a confirmation message

# **Part D – Object Usage & Output (5 Marks)**
**Write Python code to:**

1. Create two ShoppingCart objects

2. Add and remove items from both carts

3. Display individual cart details

4. Combine both carts using +

5. Display the combined cart details

```python
class ShoppingCart:
  def __init__(self,customer_name, items=None):
    self.customer_name = customer_name
    self.items = [] if items is None else items

  def add_item(self, price):
    if price > 0:
      self.items.append(price)
    else:
      print("You must provide a  positive number")

  def remove_item(self,price):
    if price in self.items:
      self.items.remove(price)
      print(f"Item price {price} removed from the cart")
    else:
      print("Item price not found in the cart")

  def calculate_total(self):
    total_amount = 0

    for item_price in self.items:
      total_amount += item_price

    return total_amount

  def apply_discount(self):

    total_amount = self.calculate_total()

    if total_amount >= 3000:
      total_amount = total_amount - (total_amount * 0.1)

    return total_amount

  def display_cart(self):
    print("-"*18 + " Cart Details " + "-"*18)
    print(f"Customer Name: {self.customer_name}")
    print(f"Items in the cart : {self.items}")
    print(f"Total Amount Purchased by the Customer : {self.calculate_total()}")
    print(f"Final Amount (After discount 10%, if Total amount >=3000): {self.apply_discount()}")
    print(f"Number of items in the cart: {len(self.items)}")

  def clear_cart(self):
    self.items.clear()
    print("All cart items are removed")

  def __add__(self,other):
    combined_Customer = self.customer_name + " & "+ other.customer_name
    combined_cart_items = self.items + other.items
    return ShoppingCart(combined_Customer,combined_cart_items)





shoppingCart1  = ShoppingCart("Abdulla Nazmus-Sakib")
shoppingCart2  = ShoppingCart("Rafiya Akter")

shoppingCart1.add_item(100)
shoppingCart1.add_item(200)
shoppingCart1.add_item(300)
shoppingCart1.add_item(400)

shoppingCart2.add_item(1000)
shoppingCart2.add_item(2000)
shoppingCart2.add_item(3000)
shoppingCart2.add_item(4000)

shoppingCart1.display_cart()
print("-"*50)
shoppingCart2.display_cart()
print("-"*50)

shoppingCart1.remove_item(100)
shoppingCart2.remove_item(1000)
print("-"*50)

print("After removing item from both cart: \n")
shoppingCart1.display_cart()
print("-"*50)
shoppingCart2.display_cart()
print("-"*50)


print("After combined both ShoppingCart: \n")
combined_cart = shoppingCart1 + shoppingCart2
combined_cart.display_cart()


print("-"*50)
shoppingCart1.clear_cart()
shoppingCart2.clear_cart()
print("-"*50)

print("After Clear all items from both cart : \n")
shoppingCart1.display_cart()
print("-"*50)
shoppingCart2.display_cart()


```
