# Question 3: User Input

Take user input for:

- Name

- Favorite programming language

Print a message:

Hello < name >, your favorite language is < language >.

```python
Name = input("Enter your name: ")
language = input("Enter your favorite programming language: ")
print("-"*50)
print(f"Hello {Name}, your favorite language is {language}.")
```

# Question 4: Type Casting (String to Integer)

1. Take two numbers from the user using input()

2. Convert them to integers

3. Print their sum

Example output:

The sum is: 30

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print("-"*50)
print(f"The sum is: {num1+num2}")
```

# Question 5: Simple Calculator

Take two numbers from the user and:

- Convert them to integers

- Print Addition

- Print Subtraction

- Print Multiplication

Example:

Addition: 20

Subtraction: 10

Multiplication: 75

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print("-"*50)
print(f"Addition: {num1+num2}")
print(f"Subtraction: {num1-num2}")
print(f"Multiplication: {num1*num2}")
```
