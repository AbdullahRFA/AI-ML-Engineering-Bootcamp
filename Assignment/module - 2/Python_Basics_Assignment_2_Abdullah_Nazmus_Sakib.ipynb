# Question 1: Loop Basics (for loop)

Write a Python program that uses a for loop to print numbers from 1 to 10.

```python
for num in range(1,11,1):
  print(num, end=" ")
```

# Question 2: Loop with Condition

Using a for loop, print only the even numbers between 1 and 20.

```python
for num in range(1,21,1):
  if num%2==0:
    print(num, end=" ")
```

# Question 3: String Manipulation

### Take a string input from the user and:

1. Print the string in uppercase

2. Print the length of the string

3. Print the string reversed

```python
inputString = input("Enter any string you want to input: ")
print("-"*50)
upperCaseString = inputString.upper()
print(f"Uppercase of the string: {upperCaseString}")
print("-"*50)
lenOfTheString = len(inputString)
print(f"Lenght of the string is: {lenOfTheString}")
print("-"*50)
reverseString = inputString[::-1]
print(f"Reversed string is: {reverseString}")
```

# Question 4: String Processing with Loop

Take a string from the user and count how many vowels (a, e, i, o, u) are in the string.

```python
inputString = input("Enter any string you want to process latter: ")

lowerCaseString = inputString.lower()
vowel_count = 0

vowel = ('a','e','i','o','u')

for char in lowerCaseString:
  if char in vowel:
    vowel_count+=1
print("-"*50)
print(f"Number of the  vowels in the inputed string: {vowel_count}")

```
